# Load Data

We will load train set and test set for our unclocked data

```python
train_path = 'data/train_data_3day_unclockedreduced1.csv' #lasso alpha = 1
test_path = 'data/test_data_unclocked.csv'
```

DataUtility is an internal tool that helps to load HFT data easier (supports creating lagged features and select a subset of reduced features)

```python
data = DataUtility()
```

Let's load train set (data & label). This load() function prints the data and label 's dimensions

```python
cols, X_train, y_train = data.load(train_path, date_index=-1, create_lag_features=False)
# this dataset already has lagged features
# cols: the columns name of returned data
# this data does not have a index (data_time) column, so we set date_index = -1
```

Let's load test set (data & label). This load() function prints the data and label 's dimensions. **This will take minutes since the test set for HFT is usually large.**

```python
_, X_test, y_test = data.load(test_path, date_index=0, create_lag_features=True, reduced_features=cols)
# (1) this dataset does not have lagged featues so we will need to create lagged features
# (2) we only select reduced features columns from train set by using reduced_features = cols
# this data has a date time index column, so we set date_index = 0
```

Let's create validation set from test set by using the first 3000 rows:

```python
nrow = 3000
X_valid = X_test[0:nrow]
y_valid = y_test[0:nrow]
X_test = X_test[nrow:]
y_test = y_test[nrow:]
```

Let's check the shape of train set, test set and validation set

```python
print("Train data shape:{}".format(X_train.shape))
print("Train label shape:{}".format(y_train.shape))
print("Test data shape:{}".format(X_test.shape))
print("Test label shape:{}".format(y_test.shape))
print("Validation data shape:{}".format(X_valid.shape))
print("Validation label shep:{}".format(y_valid.shape))
```

Let's check how unbalanced our validation set is

```python
print("Frequency for label 0, 1 and -1: {}".format(np.bincount(np.argmax(y_valid, axis = 1))))
```

# Create a DNN models using Keras/TensorFlow

```python
# Configuraitons
n_input = X_train.shape[1]
n_output = y_train.shape[1]
```

Here we use Sequential Keras model to build a 5 layers DNN. Each layer has 300 hidden units, 0.05 Ridge (L2) regularization. We use relu for activation function between layers with **batch_normalization() to normalize data**. We add a final feedforward layer with softmax activation with 3 output nodes as the probabilities for 3 classes to make sure the three probabilities will sum up to 1.0.  

There are many options of activation for hidden layers. Please refer to Keras's documentation at this (link)[https://keras.io/activations/]

```python
model = Sequential()

#layer 1
model.add(Dense(units=300, input_dim=n_input, kernel_regularizer=l2(0.05), kernel_initializer="glorot_uniform"))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(self.dropout))

#layer 2
model.add(Dense(units=300, kernel_regularizer=l2(0.05), kernel_initializer="glorot_uniform"))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(self.dropout))

# layer 3
model.add(Dense(units=300, kernel_regularizer=l2(0.05), kernel_initializer="glorot_uniform"))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(self.dropout))

# layer 4
model.add(Dense(units=300, kernel_regularizer=l2(0.05), kernel_initializer="glorot_uniform"))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(self.dropout))

# layer 5
model.add(Dense(units=300, kernel_regularizer=l2(0.05), kernel_initializer="glorot_uniform"))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(self.dropout))

model.add(Dense(n_output, activation='softmax'))
```

# Training

```python
# Configurations
n_epoch = 300
n_batch = 256
last_epoch_th = 0
initial_lrate = 0.0001
```

### Learning Rate & Optimizer

We have to also create an optimizer (a method to optimize the weights of the DNN). There are many possible options. The full list is here at this Keras documentation page: [link](https://keras.io/optimizers/)  

We will use either adagrad/adam optimization method with our desired learning rate in **lr= parameter** in this example

```python
adagrad = optimizers.Adagrad(lr=initial_lrate, epsilon=1e-08, decay=0.0)
adam = optimizers.Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
```

We then have to compile the model, using **cross entropy loss** with **adam** optimizer.  
  
A note about categorical crossentropy loss from Keras's documentation:

In order to convert integer targets into categorical targets, you can use the Keras utility to_categorical:*

### Callbacks

What is callbacks?

```python
callbacks_list = list()
```

#### Callbacks : Learning Rate scheduluer
adapt from http://machinelearningmastery.com/using-learning-rate-schedules-deep-learning-models-python-keras/

In this section, we will try to create a scheduler to change learning rate **automatically** during the training

```python
from keras.callbacks import LearningRateScheduler
import math
```

```python
# learning rate schedule
# LearningRate = InitialLearningRate * DropRate^floor(Epoch / EpochDrop)
# Or: decrease half of the learning rate after every epochs_drop (20 epochs for example)
def step_decay(epoch):
    lrate = initial_lrate
    epochs_drop = 70
    drop = 0.5
    if (epoch % epochs_drop == 0):
        lrate = initial_lrate * math.pow(drop, math.floor((1+epoch)/epochs_drop))
        print("Decreased Learning Rate to: {}".format(lrate))
    return lrate
```

```python
lrate = LearningRateScheduler(step_decay)
```

```python
callbacks_list.append(lrate)
```

#### Callbacks: Checkpoint

We will create a checkpoint to save the model weights to file everytime the validation loss improves (decreases)

```python
train_weight_path = "history/train_weights.hdf5"
checkpoint = ModelCheckpoint(train_weight_path,
        monitor='val_loss', verbose=0, save_best_only=True, mode='min')
```

```python
callbacks_list.append(checkpoint)
```

```python
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
#model.reset_states()
```

#### Data Fitting

Let's fit the model on train set

This snippet below will takes times depending on the **number of epochs (n_epoch)** and **batch size (n_batch)**  

We can interupt the traning **anytime** and **resume** anytime by changing two parameters:  
**1) initial_epoch: the epoch from which we want to resume/start**  
**2) epochs: total number of epochs we want to run**  

  
We also want to save the final model weights (the weights of the model after the last epochs) of the end of the training.

```python
history = model.fit(X_train, y_train, epochs=n_epoch, batch_size=n_batch, verbose=1,
          validation_data=(X_valid, y_valid), shuffle=True, callbacks = callbacks_list, initial_epoch = last_epoch_th)

final_weight_path = "history/final_weights.hdf5"
model.save(final_weight_path)
```

# Evaluation (on test set)
We can evaluate the model either using (1) the final weights on the DNN or (2) the best weights saved during the training. The best weights can also the same as the final weights if the final weights are the best one during the whole training.

#### (1) Evaluate using the final weights (after last epochs)

Let's evaluate the model on test set

```python
# optional (if we want to reload the weights in case the notebook is shutdown
model.load_weights(final_weight_path)
```

Let's predict the model on test set and calculate precision, recall and f1 score

```python
prediction, precision, recall, f1 = evaluate(model, X_test, y_test)
```

#### (2) Evaluate using the best weights saved during the training

First we need to load the best weights that was saved on file. 

```python
model.load_weights(train_weight_path)
```

Let's evaluate on test set using the best weights. We can **expect** that the best weights model acheives a better accuracy on the test set but little higher loss value.

Let's predict the model on test set and calculate precision, recall, and f1 score using the best weights

```python
prediction2, precision2, recall2, f12 = evaluate(model, X_test, y_test)
```

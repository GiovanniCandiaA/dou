This notebook provides the solution to the Anscombe Quartet problem in Ex. 4.9.

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
import sklearn.metrics
```

```python
np.random.seed(1)
tf.set_random_seed(1)
```

```python
anscombe_quartet = [
    {
        'x': np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]),
        'y': np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
    },
    {
        'x': np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]),
        'y': np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
    },
    {
        'x': np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]),
        'y': np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])
    },
    {
        'x': np.array([8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0]),
        'y': np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])
    }
]
```

```python
for i in range(4):
    ax = plt.subplot(2, 2, i + 1)
    ax.plot(anscombe_quartet[i]['x'], anscombe_quartet[i]['y'], 'x');
```

```python
epochs = 500
dataset_index = 1   # 0, 1, 2, or 3
```

```python
model_linear_with_hidden_layer = keras.models.Sequential()
model_linear_with_hidden_layer.add(keras.layers.Dense(1, input_dim=1, activation='linear'))
model_linear_with_hidden_layer.add(keras.layers.Dense(1, activation='linear'))
model_linear_with_hidden_layer.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
model_linear_without_hidden_layers = keras.models.Sequential()
model_linear_without_hidden_layers.add(keras.layers.Dense(1, input_dim=1, activation='linear'))
model_linear_without_hidden_layers.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
model_nonlinear = keras.models.Sequential()
model_nonlinear.add(keras.layers.Dense(3, input_dim=1, activation='tanh'))
model_nonlinear.add(keras.layers.Dense(1, activation='linear'))
model_nonlinear.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
x_train = anscombe_quartet[dataset_index]['x'].reshape((np.size(anscombe_quartet[dataset_index]['x']), 1))
y_train = anscombe_quartet[dataset_index]['y'].reshape((np.size(anscombe_quartet[dataset_index]['y']), 1))
```

```python
model_linear_with_hidden_layer.fit(x_train, y_train, epochs=epochs)
```

```python
model_linear_without_hidden_layers.fit(x_train, y_train, epochs=epochs)
```

```python
model_nonlinear.fit(x_train, y_train, epochs=epochs)
```

```python
y_train_pred_model_linear_with_hidden_layer = model_linear_with_hidden_layer.predict(x_train)
```

```python
y_train_pred_model_linear_without_hidden_layers = model_linear_without_hidden_layers.predict(x_train)
```

```python
y_train_pred_model_nonlinear = model_nonlinear.predict(x_train)
```

```python
sklearn.metrics.r2_score(y_train, y_train_pred_model_linear_with_hidden_layer)
```

```python
sklearn.metrics.r2_score(y_train, y_train_pred_model_linear_without_hidden_layers)
```

```python
sklearn.metrics.r2_score(y_train, y_train_pred_model_nonlinear)
```

```python
x_train_with_noise = x_train + np.random.normal(size=np.shape(x_train)) * np.sqrt(np.var(x_train)) / 10.
```

```python
y_train_pred_with_noise_model_linear_with_hidden_layer = model_linear_with_hidden_layer.predict(x_train_with_noise)
```

```python
y_train_pred_with_noise_model_linear_without_hidden_layers = model_linear_without_hidden_layers.predict(x_train_with_noise)
```

```python
y_train_pred_with_noise_model_nonlinear = model_nonlinear.predict(x_train_with_noise)
```

```python
plt.plot(x_train, y_train, 'x')
plt.plot(x_train, y_train_pred_model_linear_with_hidden_layer, 'x');
plt.plot(x_train_with_noise, y_train_pred_with_noise_model_linear_with_hidden_layer, 'x');
```

```python
plt.plot(x_train, y_train, 'x')
plt.plot(x_train, y_train_pred_model_linear_without_hidden_layers, 'x');
plt.plot(x_train_with_noise, y_train_pred_with_noise_model_linear_without_hidden_layers, 'x');
```

```python
plt.plot(x_train, y_train, 'x')
plt.plot(x_train, y_train_pred_model_nonlinear, 'x');
plt.plot(x_train_with_noise, y_train_pred_with_noise_model_nonlinear, 'x');
```
